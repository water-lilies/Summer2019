## 데이타프레임의 병합
- 두 데이터 프레임의 공통 컬럼이나 인덱스를 기준으로 두 개의 테이플을 합친다.
- 이 때 기준이 되는 컬럼과 인덱스를 키(key)라고 한다.
```
pd.merge(데이타프레임1, 데이타프레임2) : Inner Join
pd.merge(데이타프레임1, 데이타프레임2, how='outer/left/right')
```

````python
df1 = pd.DataFrame({
    '고객번호' : [1001,1002,1003,1004,1005,1006,1007],
    '이름' : ['둘리','도우너','또치','길동','희동','마이콜','영희']
}, columns=['고객번호','이름'])
df1
````

````python
df2 = pd.DataFrame({
    '고객번호' : [1001,1001,1005,1006,1008,1001],
    '금액' : [10000,20000,15000,5000,100000,30000]
}, columns=['고객번호','금액'])
df2
````

## inner join
- 
- 양쪽 데이타프레임에 모두 키가 존재하는 데이터만 보여준다
pd.merge(데이타프레임1,데이타프레임)

````python
pd.merge(df1,df2)
````

## outer join방식으로 병합하기

````python
pd.merge(df1,df2, how='outer')
````

## right join 방식으로 병합하기
- 두번째 데이터프레임의 인덱스 카값을 모두 보여준다

````python
pd.merge(df1,df2,how='right')
````

## 인덱스를 기준으로 컬럼 합치기
pd.merge(데이타프레임1,데이타프레임2, left_index=True, right_index=True)

````python
pd.merge(df1,df2,left_index=True, right_index=True)
````
