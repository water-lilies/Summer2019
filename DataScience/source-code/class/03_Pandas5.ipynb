```python
import numpy as np
import pandas as pd
```

```python
practice = pd.DataFrame({
    '날짜':[],
    '운동':[],
    '시간':[]
})
practice
```

## 날짜별/운동별 데이터를 한눈에 보기 좋게 표로 표시해 보십시오.

```python
practice.pivot('날짜','운동','시간')
```

```python
prac_pivot = practice.pivot('날짜','운동','시간')
prac_pivot.fillna(0, inplace=True)
prac_pivot
```

```python
practice.pivot_table(index='운동')
```

```python
practice.pivot_table(index='운동', aggfunc=np.sum) #기본값이 np.mean
```

```python
practice.pivot_table(index='운동', aggfunc=[np.sum, len])
```

## pivot_table()

- 데이터 열 중에서 두 개의 열을 각각 행 인덱스, 열 인덱스로 사용하여 데이터를 조회하여 펼쳐놓은 것

## pivot()

- 첫번째 인수 : 행 인덱스로 사용할 열 이름
- 두번째 인수 : 열 인덱스로 사용할 열 이름
- 마지막 인수 : 데이터로 사용할 열 이름
