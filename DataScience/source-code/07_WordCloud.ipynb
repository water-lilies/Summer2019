```python
wc = WordCloud(background_color='white', max_font_size=200, 
               font_path ='C:\Windows\Fonts\malgunbd.ttf' ,stopwords = STOPWORDS)
wc = wc.generate(text)
wc.words_
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wc)
# 크기와 관련된 옵션
plt.tight_layout(pad=0)
plt.axis('off')
# 저장
plt.savefig('../image/file.png',bbox_inches='tight')
```

### 배경이미지

```python
import numpy as np
from PIL import Image

text = open('../data/wordcloud/alice.txt').read()
heart_mask = np.array(Image.open('../data/wordcloud/heart.jpg'))
stopword = set(STOPWORDS)
stopword.add("said")

wc = WordCloud(background_color='white', 
               max_words=200, mask=heart_mask, stopwords=stopword)
wc = wc.generate(text)

plt.figure(figsize=(10,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show
```
